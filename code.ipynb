```python
import pandas as pd

# Load the data
usa_gdp = pd.read_csv('USA_GDP.csv')
india_gdp = pd.read_csv('India_GDP_Data1.csv')

# Clean the data (if necessary)
# usa_gdp = usa_gdp.dropna()
# india_gdp = india_gdp.dropna()

# Analyze the data
print('USA GDP Statistics:')
print(usa_gdp.describe())
print('\nIndia GDP Statistics:')
print(india_gdp.describe())

# Compare the GDPs
comparison = pd.DataFrame({
    'Year': usa_gdp['Year'],
    'USA_GDP': usa_gdp['GDP'],
    'India_GDP': india_gdp['GDP']
})
print('\nGDP Comparison:')
print(comparison)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
usa_data = pd.read_csv('USA_GDP.csv')
india_data = pd.read_csv('India_GDP_Data1.csv')

plt.plot(usa_data['Year'], usa_data['GDP'], label='USA')

plt.plot(india_data['Year'], india_data['GDP'], label='India')


plt.xlabel('Year')
plt.ylabel('GDP (in billions)')
plt.title('GDP Analysis: USA vs India')


plt.legend()


plt.grid()
plt.show()
```
